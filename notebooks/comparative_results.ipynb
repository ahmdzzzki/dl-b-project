```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for comparative results
models = [
    "Bidirectional LSTM with Attention",
    "Stacked LSTM with SMOTE and Regularization",
    "FCNN with SMOTETomek",
    "FCNN with SMOTE"
]
pr_auc = [0.7831, 0.8326, 0.7989, 0.8063]
precision = [43, 50, 47, 46]
recall = [91, 92, 91, 91]
f1_score = [59, 64, 62, 61]
```

Bar chart for PR AUC

```python
plt.figure(figsize=(10, 6))
plt.bar(models, pr_auc, color='skyblue', edgecolor='black')
plt.title("PR AUC Comparison", fontsize=16)
plt.ylabel("PR AUC", fontsize=12)
plt.xticks(rotation=15, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('../figures/pr_auc_comparison.png', dpi=300)
plt.show()
```

Bar chart for Precision, Recall, and F1-Score

```python
x = np.arange(len(models)) 
width = 0.25 

fig, ax = plt.subplots(figsize=(12, 6))
rects1 = ax.bar(x - width, precision, width, label='Precision', color='lightcoral')
rects2 = ax.bar(x, recall, width, label='Recall', color='lightgreen')
rects3 = ax.bar(x + width, f1_score, width, label='F1-Score', color='lightblue')

ax.set_ylabel('Percentage (%)', fontsize=12)
ax.set_title('Comparison of Precision, Recall, and F1-Score', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=15, ha='right', fontsize=10)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

fig.tight_layout()
plt.savefig('../figures/precision_recall_f1_comparison.png', dpi=300)
plt.show()
```
